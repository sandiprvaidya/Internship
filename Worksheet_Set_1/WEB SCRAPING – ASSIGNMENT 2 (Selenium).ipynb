```python
# Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
# have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
# jobs data.

driver.get("https://www.shine.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
title_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]/div/h2')
for i in title_tags:
    title= i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags:
    location= i.text
    job_location.append(location)
    
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags:
    company= i.text
    company_name.append(company)    
    
experience_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags:
    experience= i.text
    experience_required.append(experience)    
```

```python
print(len(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python

df=pd.DataFrame({'Title':job_title[0:10],'Location':job_location[0:10],'Company_Name':company_name[0:10],'Experience':experience_required[0:10]})
df
```

```python
# Scrape data of first 100 sunglasses listings on flipkart.com.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/")
```

```python
item = driver.find_element(By.CLASS_NAME,"Pke_EE")
item.send_keys('sunglasses')
```

```python
sunglass_brand=[]
product_details=[]
sunglass_price=[]
```

```python
start= 0
end= 4
for page in range(start,end):
    sunglass_name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in sunglass_name:
        brand= i.text
        sunglass_brand.append(brand)
    
    product_info = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in product_info:
        product= i.text
        product_details.append(product) 
    
    sunglass_cost = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in sunglass_cost:
        price= i.text
        sunglass_price.append(price) 
     
    next_button = driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(4)    
```

```python
print(len(sunglass_brand),len(product_details),len(sunglass_price))
```

```python
df=pd.DataFrame({'Brand':sunglass_brand[0:100],'Product Description':product_details[0:100],'Price':sunglass_price[0:100]})
df
```

```python
 # scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
 # 1. Title 2. Ratings 3. Price


import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(" https://www.amazon.in/")
```

```python
item = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
item.send_keys('Laptop')
```

```python
laptop_title=[]
laptop_ratings=[]
laptop_price=[]
```

```python
title_tags = driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in title_tags:
    title= i.text
    laptop_title.append(title)
    
product_ratings = driver.find_elements(By.XPATH,'//i[@class="a-icon a-icon-star-small a-star-small-4 aok-align-bottom"]')
for i in product_ratings:
    ratings= i.text
    laptop_ratings.append(ratings)
    
product_cost = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in product_cost:
    price= i.text
    laptop_price.append(price) 
```

```python
print(len(laptop_title),len(laptop_ratings),len(laptop_price))
```

```python
df=pd.DataFrame({'Title':laptop_title[0:10],'Ratings':laptop_ratings[0:10],'Price':laptop_price[0:10]})
df
```

```python
# Write a python program to scrape data for Top 1000 Quotes of All Time.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
book_title=[]
book_author=[]
book_quotes=[]
```

```python
start= 0
end= 10
for page in range(start,end):
    book_name = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in book_name:
        book_title.append(i.text)
    
    book_writer = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in book_writer:
        book_author.append(i.text) 
    
    book_headlines = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in book_headlines:
        book_quotes.append(i.text) 
     
    next_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]')
    next_button.click()
    time.sleep(10)    
```

```python
print(len(book_title),len(book_author),len(book_quotes))
```

```python
df=pd.DataFrame({'Quote':book_title[0:1000],'Author':book_author[0:1000],'Types of Quotes':book_quotes[0:1000]})
df
```
